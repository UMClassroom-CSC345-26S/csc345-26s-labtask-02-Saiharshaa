```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

# Load dataset
df = pd.read_csv('AllCars.csv')

# [cite_start]Preprocessing: Remove non-ordinal 'Make' [cite: 1]
# Features: Volume, Doors. Target: Style
X = df[['Volume', 'Doors']]
y = df['Style']

# Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# [cite_start]Split 80/20 [cite: 1]
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Iterate K to find highest accuracy
accuracies = []
ks = range(1, 21)
best_k = 1
max_acc = 0

for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    acc = knn.score(X_test, y_test)
    accuracies.append({'K': k, 'Accuracy': acc})
    if acc > max_acc:
        max_acc = acc
        best_k = k

# [cite_start]Save Accuracy.csv [cite: 1]
pd.DataFrame(accuracies).to_csv('Accuracy.csv', index=False)

# Re-train best model
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train, y_train)

# [cite_start]Prep Testing.csv with Prediction and Confidence [cite: 1]
test_df = pd.DataFrame(scaler.inverse_transform(X_test), columns=['Volume', 'Doors'])
test_df['Style'] = y_test.values
test_df['Prediction'] = best_knn.predict(X_test)
test_df['Confidence'] = np.max(best_knn.predict_proba(X_test), axis=1)

# Save Final Files
test_df.to_csv('Testing.csv', index=False)
train_df = pd.DataFrame(scaler.inverse_transform(X_train), columns=['Volume', 'Doors'])
train_df['Style'] = y_train.values
train_df.to_csv('Training.csv', index=False)

print(f"Optimal K: {best_k} with Accuracy: {max_acc}")
```
